# Performance Benchmarking of LCS Algorithms

This report presents the performance benchmarking of several Longest Common Subsequence (LCS) algorithms. The study compares these algorithms on different input sizes to evaluate their time and space complexity.

1. recursive
2. memoized
3. dynamic programming (DP)
4. optimized DP
5. reconstruction

We'll analyze:
- Time complexity
- Space complexity
- i added the comparison compilxity

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

# Add parent directory to path for importing complexity package
sys.path.append(str(Path.cwd().parent))

from complexity.analyser import (
    ComplexityAnalyzer,
    ConstantComplexity,
    LogarithmicComplexity,
    LinearComplexity,
    LinearLogComplexity
)

from complexity.visualizer import ComplexityVisualizer

import seaborn as sns
sns.set_theme()
sns.set_palette('husl')
```

```python
# Load the data
raw_df = pd.read_csv('../benchmark_results.csv')

print("Raw data shape:", raw_df.shape)
print("\nColumns:", raw_df.columns.tolist())
```

```python
# Create analyzer with specific complexity functions
complexity_functions = [
    ConstantComplexity(),
    LinearComplexity(),
    LogarithmicComplexity(),
    LinearLogComplexity()
]
analyzer = ComplexityAnalyzer(complexity_functions)

# Analyze complexity for each algorithm
print("\nAnalyzing algorithmic complexity...")

for func_name in raw_df['function'].unique():
    x = raw_df['size'].values
    time_y = raw_df['time'].values
    comp_y = raw_df['comparison_count'].values
    memory_y = raw_df['memory'].values
    
    # Create visualizers
    time_visualizer = ComplexityVisualizer(x, time_y, complexity_functions)
    comp_visualizer = ComplexityVisualizer(x, comp_y, complexity_functions)
    memory_visualizer = ComplexityVisualizer(x, memory_y, complexity_functions)
    
    time_complexity, _ = analyzer.get_best_fit(x, time_y)
    comp_complexity, _ = analyzer.get_best_fit(x, comp_y)
    memory_complexity, _ = analyzer.get_best_fit(x, memory_y)
    
    # Plot the results
    time_visualizer.plot(
        time_complexity,
        title=f"{func_name} - Time Complexity"
    )
    
    comp_visualizer.plot(
        comp_complexity,
        title=f"{func_name} - comp Complexity"
    )
    memory_visualizer.plot(
        memory_complexity,
        title=f"{func_name} - space Complexity"
    )
    
    print(f"Time Complexity: {time_complexity}")
    print(f"comp Complexity: {comp_complexity}")
    print(f"memory Complexity: {memory_complexity}")
    
    
```
